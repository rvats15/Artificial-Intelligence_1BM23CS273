```python
import math

# --------------------------------------------------
# CLEAN & FULLY WORKING ALPHA-BETA PRUNING
# --------------------------------------------------

def alpha_beta(depth, index, maximizingPlayer, values, alpha, beta, max_depth, pruned):
    # leaf node
    if depth == max_depth:
        return values[index]

    if maximizingPlayer:
        best = -math.inf
        # Each node has 2 children: left = 2*index, right = 2*index + 1
        for child in [2*index, 2*index + 1]:
            val = alpha_beta(depth + 1, child, False, values, alpha, beta, max_depth, pruned)
            best = max(best, val)
            alpha = max(alpha, best)

            if beta <= alpha:
                pruned.append((index, child))
                break

        return best

    else:
        best = math.inf
        for child in [2*index, 2*index + 1]:
            val = alpha_beta(depth + 1, child, True, values, alpha, beta, max_depth, pruned)
            best = min(best, val)
            beta = min(beta, best)

            if beta <= alpha:
                pruned.append((index, child))
                break

        return best


# -----------------------------------------
# SAMPLE DATA (leaf nodes)
# -----------------------------------------
# Tree in array form (like AI Lab example)
values = [3, 5, 6, 9, 1, 2, 0, -1]
max_depth = 3        # depth of leaf nodes
pruned = []          # store pruned nodes

# -----------------------------------------
# CALL FUNCTION
# -----------------------------------------
result = alpha_beta(
    depth=0,
    index=0,
    maximizingPlayer=True,
    values=values,
    alpha=-math.inf,
    beta=math.inf,
    max_depth=max_depth,
    pruned=pruned
)

print("Optimal Value at Root =", result)
print("Pruned Branches =", pruned)
```
